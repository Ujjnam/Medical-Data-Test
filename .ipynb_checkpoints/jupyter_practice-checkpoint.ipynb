```python
import string
sent = 'ignited minds'
res = ''.join(ch for ch in sent if  (ch not in string.punctuation + string.whitespace))

dit = {}
for ch in res:
    if(ch in dit):
        dit[ch] = dit[ch] + 1
    else:
        dit[ch] = 1
print(dit)        
```

```python
word = 'then'
rev = ''
count = len(word)-1
while(count>=0):
    rev = rev + word[count]
    count -= 1
print(rev)    
```

```python
import pandas as pd
import oracledb

conn_str = 'myuser/mypass123@localhost:1521/xepdb1'
conn=cur=None
try:
    conn = oracledb.connect(conn_str)
    cur = conn.cursor()
    df = pd.read_excel('medical_data.xlsx')
    df['ADMISSION_DATE'] = df['ADMISSION_DATE'].dt.strftime('%d-%m-%Y')
    df['DISCHARGE_DATE'] = df['DISCHARGE_DATE'].dt.strftime('%d-%m-%Y')
    df['BILLING_AMOUNT'] =  df['BILLING_AMOUNT'].apply(lambda x : f"{x:.2f}")
    data_to_insert = df.to_records(index=False).tolist()
    query = """ INSERT INTO MEDICAL_DATA (PATIENT_ID,PATIENT_NAME,AGE,GENDER,BLOOD_TYPE,
                                          MEDICAL_CONDITION,DOCTOR,HOSPITAL,BILLING_AMOUNT
                                          ,ADMISSION_TYPE,ADMISSION_DATE,DISCHARGE_DATE) VALUES
                                          (:PATIENT_ID,:PATIENT_NAME,:AGE,:GENDER,:BLOOD_TYPE,:MEDICAL_CONDITION,:DOCTOR,:HOSPITAL,
                                          :BILLING_AMOUNT,:ADMISSION_TYPE,TO_DATE(:ADMISSION_DATE,'DD-MM-YYYY'),TO_DATE(:DISCHARGE_DATE,'DD-MM-YYYY'))"""
    cur.executemany(query,data_to_insert)
    conn.commit()
except Exception as err:
    print('Unable to connect to DB.',err)
finally:
    if(cur):
        cur.close()
    if(conn):
        conn.close()
```
