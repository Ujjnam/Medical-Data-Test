```python
import re

file = open('info.txt','r')
data = file.readline().strip()
count = 0
while(data):
    count += len(re.findall(r'\band\b',data,re.IGNORECASE))
    data = file.readline().strip()
print(count)    
file.close()    
```

```python
import re
import pandas as pd
file = open('sample.txt','r')
data = file.readlines()
words = ['AI','THE','ARE']
diction = {} # or diction = {word:0 for word in words}
for line in data:
    for word in words:
        count = len(re.findall(rf'\b{word}\b',line,re.IGNORECASE))
        if word not in diction:
            diction[word]  = count
        else:
            diction[word]  = diction[word] + count
print(diction)
df = pd.DataFrame(list(diction.items()),columns=['word','count'])
df.to_excel('words_count.xlsx')
file.close()
```
