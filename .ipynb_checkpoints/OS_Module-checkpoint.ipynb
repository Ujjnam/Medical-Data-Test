```python
import os

source_path = r'.\BILLS'
for item in os.listdir(source_path):
    item_path = os.path.join(source_path,item)
    if (os.path.isfile(item_path)):
        if(item.endswith('23.txt')):
            year_path = '2023' 
            target_path = os.path.join(source_path,year_path,item)
            os.rename(item_path,target_path) 
        elif(item.endswith('24.txt')):
            year_path = '2024' 
            target_path = os.path.join(source_path,year_path,item)
            os.rename(item_path,target_path)  
        elif(item.endswith('25.txt')):
            year_path = '2025' 
            target_path = os.path.join(source_path,year_path,item)
            os.rename(item_path,target_path)     
```

```python
import os
file = open('order_log.txt','r')
data = file.readlines()
order_count = 0
amount = 0
for line in data:
    order_count += 1
    amount += int(line.strip().split(';')[3].split(':')[1])
print(f'Total orders = {order_count}')
print(f'Total amount = {amount}')
file.close()
```

```python
import os
file = open('order_log.txt','r')
data = file.readlines()
mumbai_orders = 0
mumbai_amount = 0
for order in data:
    city = order.strip().split(';')[2].split(':')[1];
    if(city.upper() == 'MUMBAI'):
        mumbai_orders += 1
        mumbai_amount += int(order.strip().split(';')[3].split(':')[1])
print(mumbai_orders)
print(mumbai_amount)
file.close()    
```
