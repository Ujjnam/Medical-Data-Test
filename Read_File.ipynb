```python
f = open('info.txt','r')
#READING ENTIRE FILE WILL AFFECT RAM USAGE IF RECORDS ARE HUGE
#HENCE READ LINE BY LINE
line = f.readline() # now cursor moved to 2nd line
while (line):
    print(len(line.split()))
    print(line)
    line = f.readline() # line variable has 2nd line
f.close()
```

```python
#WRITING TO OTHER FILE
read_file = open('info.txt','r')
write_file = open('info_write.txt','w')

line = read_file.readline()
while(line):
    write_file.write(line)
    line = read_file.readline()
read_file.close()
write_file.close()
```
